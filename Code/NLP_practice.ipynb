```python
import nltk
from nltk import CFG
from nltk.tokenize import word_tokenize
from nltk.parse.generate import generate

def parse_sent(sent):
    pos_tokens = nltk.pos_tag(word_tokenize(sent))

    # Initialize token holders
    tok_NNP = None
    tok_V = None
    tok_NN = None

    for word, pos in pos_tokens:
        if pos == "NNP" and not tok_NNP:
            tok_NNP = f"'{word}'"
        elif pos in ["VBD", "VBN", "VBG"] and not tok_V:
            tok_V = f"'{word}'"
        elif pos == "NN" and not tok_NN:
            tok_NN = f"'{word}'"

    if not all([tok_NNP, tok_V, tok_NN]):
        print("Sentence doesn't contain required POS structure (NNP + Verb + NN)")
        return

    # Define CFG with populated terminals
    grammar = CFG.fromstring(f"""
        S -> NP VP
        NP -> NNP
        VP -> V NN
        NNP -> {tok_NNP}
        V -> {tok_V}
        NN -> {tok_NN}
    """)

    # Generate and print all possible sentences
    for sentence in generate(grammar):
        print(' '.join(sentence))

# Example call
parse_sent("John ate cake")
```

## Bag of words

```python
import nltk
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
```

```python
def extract_words(sentance):
    words=nltk.word_tokenize(sentance)
    cleared_text =[w.lower() for w in words if w not in stop_words]
    return cleared_text

def tokanize_sentance(sentances):
    words = []
    for sentance in sentances:
        w= extract_words(sentance)
        words.extend(w)
    words= sorted(list(set(words)))
    return words

def bag_of_words(sentance, words):
    sentance_words = extract_words(sentance)
    #frequency word count
    bag = np.zeros(len(words))

    for sw in sentance_words:
        for i,word in enumerate(words):
            if word == sw:
                bag[i] += 1
    return np.array(bag)
```

```python
corpus = ["John saw the train"
"The train was late"
"Max and Rob took the bus"
"I looked for Max and Rob at the bus station",
"Max and Rob arrived at the bus station early but waited until noon for the bus"]
vacab = tokanize_sentance(corpus)
print(vacab)
```

```python
bag_of_words("Max and Rob took the bus", vacab) # word count vector
```

```python
bag_of_words("max and rob arrived at the bus station early but waited until noon for the bus", vacab) # word count vector
```

```python
bag_of_words("Jhon saw the train", vacab) # word count vector
```

### Converting Text to Features and Labels

```python
import pandas as pd
import numpy as np
import random
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
from nltk.stem import WordNetLemmatizer
stemmer = WordNetLemmatizer()
```

```python
dataset= pd.read_csv("fake_or_real_news.csv", encoding='latin-1')
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
import re
from nltk.stem import WordNetLemmatizer

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

corpus = []

for i in range(0, 6335):
    # Remove non-word characters (anything other than letters and numbers)
    document = re.sub(r'\W', ' ', dataset['text'][i])
    
    # Remove single characters
    document = re.sub(r'\s+[a-zA-Z]\s+', ' ', document)
    
    # Remove single characters from the start
    document = re.sub(r'^[a-zA-Z]\s+', '', document)
    
    # Substitute multiple spaces with a single space
    document = re.sub(r'\s+', ' ', document, flags=re.I)
    
    # Convert to lowercase
    document = document.lower()
    
    # Tokenize
    document = document.split()
    
    # Lemmatize each word
    document = [lemmatizer.lemmatize(word) for word in document]
    
    # Join back into one string
    document = ' '.join(document)
    
    corpus.append(document)
```

```python
print(corpus[0])
```

```python
X= corpus
y= dataset.label
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Use built-in English stop words
count_vectorizer = CountVectorizer(max_features=1500, min_df=5, max_df=0.7, stop_words='english')

# Fit-transform on training data
count_train = count_vectorizer.fit_transform(X_train)

# Transform test data
count_test = count_vectorizer.transform(X_test)
```

```python
print(count_vectorizer.get_feature_names_out())
```

```python
pd.DataFrame(count_train.toarray(), columns=count_vectorizer.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1500, min_df=5, max_df=0.7, stop_words='english')
tfidf_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_test = tfidf_vectorizer.transform(X_test)
```
